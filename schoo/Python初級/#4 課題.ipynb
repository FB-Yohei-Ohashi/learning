```python
# pandasで表を取得する
url = "https://www.data.jma.go.jp/stats/data/mdrr/synopday/data1s.html"

# html内の全ての表を取得するので必要な1-5の表に絞る
get_dfs = pd.read_html(url)
# lenすればいくつデータフレームがあるか確認できる
#print(len(get_dfs)) # 9

# 今回使いたいデータフレームを取得
target_dfs = get_dfs[0:5]
```
