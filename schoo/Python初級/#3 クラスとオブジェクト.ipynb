## クラスとオブジェクト

### シンプルなクラス

```python
class ExercisePerson:
    def __init__(self, name: str, weight: int, height: int, hp: int) -> None:
        print("初期化メソッドが実行されました")
        self._name = name
        self._weight = weight
        self._height = height
        self._hp = hp
    
    @property
    def name(self) -> str:
        return self._name
    
    @property
    def weight(self) -> int:
        return self._weight
    
    @weight.setter
    def weight(self, value: int) -> None:
        self._weight = value
    
    @property
    def height(self) -> int:
        return self._height
    
    @property
    def hp(self) -> int:
        return self._hp
    
    @hp.setter
    def hp(self, value: int) -> None:
        self._hp = value
    
    
    def exercise(self) -> None:
        print(f"{self.name}さんは運動したので1kg体重が減りました！")
        self.weight -= 1
    
    def eat(self) -> None:
        print(f"{self.name}さんは食事したので1kg体重が増えました！")
        self.weight += 1
    
    def sleep(self) -> None:
        self.hp += 1
    
    def work(self) -> None:
        self.hp -= 1
```

```python
person1 = ExercisePerson("Ken", 100, 180, 100)
person2 = ExercisePerson("Emi", 50, 160, 100)
```

```python
print(person1.name)
print(f"{person1.weight}kg")
print(f"{person1.height}cm")
person1.exercise()
person1.exercise()
print(f"{person1.weight}kg")
person1.work()
person1.work()
print(f"HP: {person1.hp}")
```

```python
print(person2.name)
print(f"{person2.weight}kg")
print(f"{person2.height}cm")
person2.eat()
print(f"{person2.weight}kg")
person2.sleep()
person2.sleep()
print(f"HP: {person2.hp}")
```
